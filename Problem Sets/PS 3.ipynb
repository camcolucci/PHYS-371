```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 0
L = 1
A = 1

# Standing wave function for each mode
def standing_wave(n, x):
    k = (n + 0.5) * np.pi / L  # Calculating k for each mode
    return A * np.sin(k * x)   # Since B=0, we only use the sine term

# Define x values across the length of the string
x = np.linspace(a, L, 500)

# Plotting the first three modes
plt.figure(figsize=(8, 6))
for n in range(3):
    y = standing_wave(n, x)
    plt.plot(x, y, label=f'Mode {n + 1}')

# Adding labels and title
plt.xlabel('Position along string (x)')
plt.ylabel('Displacement (y)')
plt.title('First Three Standing Wave Modes on a Fixed-Free String')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def initial_condition(x):
    return np.piecewise(x, [x < L/3, (x >= L/3) & (x <= 2*L/3), x > 2*L/3], [lambda x: 3*x/L, 1, lambda x: -3*x/L + 3])

x = np.linspace(0, L, 500)
y = initial_condition(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\xi(x, t=0)$', color='b')
plt.xlabel('Position along string (x)')
plt.ylabel(r'$\xi(x, t=0)$')
plt.title('Initial Conditions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1        # Total length of the string
A = 1        # Amplitude of oscillation
v_left = 1   # Phase speed on left half (String A) as v0
v_right = 2  # Phase speed on right half (String B) as 2v0

# Frequency calculation based on the left side (String A)
def frequency(n):
    return (v_left / L) * (n + 0.5)

# Standing wave function for each half
def standing_wave_left(n, x_left):
    f = frequency(n)  # Use the same frequency for both halves
    k_left = 2 * np.pi * f / v_left  # Wavenumber for left half
    return A * np.sin(k_left * x_left)

def standing_wave_right(n, x_right):
    f = frequency(n)  # Use the same frequency for both halves
    k_right = 2 * np.pi * f / v_right  # Wavenumber for right half
    return A * np.sin(k_right * (L - x_right))

# Define x values for the left and right halves
x_left = np.linspace(0, L / 2, 250)
x_right = np.linspace(L / 2, L, 250)

# Plotting the first three modes and printing their frequencies (same frequency for both sides)
plt.figure(figsize=(10, 6))
for n in range(3):
    f = frequency(n)
    print(f"Mode {n + 1} - Frequency: {f:.2f} Hz (Same for both halves)")

    y_left = standing_wave_left(n, x_left)
    y_right = standing_wave_right(n, x_right)
    
    # Combine the left and right halves
    x_combo = np.concatenate((x_left, x_right))
    y_combo = np.concatenate((y_left, y_right))
    
    plt.plot(x_combo, y_combo, label=f'Mode {n + 1} (f = {f:.2f} Hz)')

# Adding labels and title
plt.xlabel('Position along string (x)')
plt.ylabel('Displacement (y)')
plt.title('First Three Standing Wave Modes with Center Node and Different Phase Speeds')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1        # Total length of the string
A = 1        # Amplitude of oscillation
v_left = 1   # Phase speed on left half (String A) as v0
v_right = 1.5  # New phase speed on right half (String B) as (3v0 / 2)

# Standing wave function for each half
def standing_wave_left(n, x_left):
    f = frequency(n)
    k_left = 2 * np.pi * f / v_left  # Wavenumber for left half
    return A * np.sin(k_left * x_left)

def standing_wave_right(n, x_right):
    f = frequency(n)
    k_right = 2 * np.pi * f / v_right  # Wavenumber for right half (with updated v_right)
    return A * np.sin(k_right * (L - x_right))  # Adjust to reflect the right half

# Define x values for the left and right halves
x_left = np.linspace(0, L / 2, 250)
x_right = np.linspace(L / 2, L, 250)

# Plotting the first three modes and printing their frequencies
plt.figure(figsize=(10, 6))
for n in range(3):
    f = frequency(n)
    print(f"Mode {n + 1} frequency with updated v_right: {f:.2f} Hz")

    y_left = standing_wave_left(n, x_left)
    y_right = standing_wave_right(n, x_right)
    
    # Combine the left and right halves
    x_combo = np.concatenate((x_left, x_right))
    y_combo = np.concatenate((y_left, y_right))
    
    plt.plot(x_combo, y_combo, label=f'Mode {n + 1} (f = {f:.2f} Hz)')

# Adding labels and title
plt.xlabel('Position along string (x)')
plt.ylabel('Displacement (y)')
plt.title('First Three Standing Wave Modes with Updated Phase Speed on Right Half')
plt.legend()
plt.grid(True)
plt.show()
```

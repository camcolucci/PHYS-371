```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 0
L = 1
A = 1

# Standing wave function for each mode
def standing_wave(n, x):
    k = (n + 0.5) * np.pi / L  # Calculating k for each mode
    return A * np.sin(k * x)   # Since B=0, we only use the sine term

# Define x values across the length of the string
x = np.linspace(a, L, 500)

# Plotting the first three modes
plt.figure(figsize=(8, 6))
for n in range(3):
    y = standing_wave(n, x)
    plt.plot(x, y, label=f'Mode {n + 1}')

# Adding labels and title
plt.xlabel('Position along string (x)')
plt.ylabel('Displacement (y)')
plt.title('First Three Standing Wave Modes on a Fixed-Free String')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def initial_condition(x):
    return np.piecewise(x, [x < L/3, (x >= L/3) & (x <= 2*L/3), x > 2*L/3], [lambda x: 3*x/L, 1, lambda x: -3*x/L + 3])

x = np.linspace(0, L, 500)
y = initial_condition(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\xi(x, t=0)$', color='b')
plt.xlabel('Position along string (x)')
plt.ylabel(r'$\xi(x, t=0)$')
plt.title('Initial Conditions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_points(N, R):
    # Generate N random points within the square
    x = np.random.uniform(-R, R, N)
    y = np.random.uniform(-R, R, N)
    return x, y

def estimate_pi(N, R):
    x, y = generate_points(N, R)  # Generate random points
    inside_circle = x**2 + y**2 < R**2  # Check if points are inside the circle
    N_c = np.sum(inside_circle)  # Count the number of points inside the circle
    pi_estimate = 4 * N_c / N  # Estimate the value of π
    return pi_estimate

# Define a range of N values
N_values = [100, 1000, 5000, 10000, 50000, 100000, 500000]
R = 1
pi_estimates = []
differences = []

# Calculate π estimate and differences for each N value
for N in N_values:
    pi_estimate = estimate_pi(N, R)
    pi_estimates.append(pi_estimate)
    difference = pi_estimate - np.pi  # Difference between estimated and actual π
    differences.append(difference)
    print(f"N = {N}, Estimated π: {pi_estimate}, Difference: {difference}")

# Plotting the differences as a function of N
plt.figure(figsize=(10, 6))
plt.plot(N_values, differences, marker='o', linestyle='-', label='Difference (Estimated π - Actual π)')
plt.axhline(y=0, color='r', linestyle='--', label='Zero Difference Line')
plt.xscale('log')  # Logarithmic scale for better visualization
plt.xlabel('Number of Points (N)')
plt.ylabel('Difference from Actual π')
plt.title('Difference between Estimated and Actual π as N Increases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to be integrated
def f(x):
    return np.exp(-np.sqrt(x))

# Monte Carlo integration method
def monte_carlo_integration(N):
    # Generate N random x-values between 0 and 1
    x = np.random.uniform(0, 1, N)
    # Generate N random y-values between 0 and 1
    y = np.random.uniform(0, 1, N)
    # Count how many points fall below the curve y < f(x)
    points_below_curve = y < f(x)
    # Estimate the integral as the ratio of points below the curve
    I_estimate = np.sum(points_below_curve) / N
    return I_estimate

# Number of random points to use
N = 100000
# Estimate the integral
I_estimate = monte_carlo_integration(N)
print(f"Estimated value of the integral I: {I_estimate}")

# Plotting the function and random points for visualization
x_vals = np.linspace(0, 1, 1000)
y_vals = f(x_vals)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = e^{-\sqrt{x}}$', color='blue')
# Plot random points
plt.scatter(np.random.uniform(0, 1, 100), np.random.uniform(0, 1, 100), s=10, color='red', alpha=0.5, label='Random Points')
plt.fill_between(x_vals, 0, y_vals, color='blue', alpha=0.2, label='Area under curve')
plt.title('Monte Carlo Integration Visualization')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f(x):
    return np.exp(-np.sqrt(x))

# Monte Carlo integration function
def monte_carlo_integration(N):
    # Generate N random x-values between 0 and 1
    x = np.random.uniform(0, 1, N)
    # Generate N random y-values between 0 and 1
    y = np.random.uniform(0, 1, N)
    # Count points where y is less than f(x) to determine points below the curve
    points_below_curve = y < f(x)
    # Estimate the integral as the fraction of points below the curve
    I_estimate = np.sum(points_below_curve) / N
    return I_estimate

# Define a range of N values to test
N_values = [100, 500, 1000, 5000, 10000, 50000, 100000, 500000]
integral_estimates = []

# Evaluate the integral for each N
for N in N_values:
    I_estimate = monte_carlo_integration(N)
    integral_estimates.append(I_estimate)
    print(f"Estimated value of I for N = {N}: {I_estimate}")

# Plotting the estimated integral values as a function of N
plt.figure(figsize=(10, 6))
plt.plot(N_values, integral_estimates, marker='o', linestyle='-', label='Estimated Integral')
plt.axhline(y=0.52848, color='r', linestyle='--', label='Actual Integral Value (Approx)')
plt.xscale('log')  # Use logarithmic scale for N values
plt.xlabel('Number of Points (N)')
plt.ylabel('Estimated Integral I')
plt.title('Estimation of Integral I as N Increases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to be integrated
def f(x):
    return np.exp(-np.sqrt(x))

# Trapezoidal rule integration function
def trapezoidal_integration(N):
    a = 0  # Lower bound of the integral
    b = 1  # Upper bound of the integral
    h = (b - a) / N  # Width of each slice
    x = np.linspace(a, b, N + 1)  # N + 1 points including endpoints
    y = f(x)  # Function values at these points
    I_estimate = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])  # Trapezoidal rule formula
    return I_estimate

# Define a range of N values (number of slices)
N_values = [10, 50, 100, 500, 1000, 5000, 10000, 50000]
integral_estimates = []

# Evaluate the integral using the trapezoidal rule for each N
for N in N_values:
    I_estimate = trapezoidal_integration(N)
    integral_estimates.append(I_estimate)
    print(f"Estimated value of I for N = {N}: {I_estimate}")

# Plotting the estimated integral values as a function of N
plt.figure(figsize=(10, 6))
plt.plot(N_values, integral_estimates, marker='o', linestyle='-', label='Estimated Integral (Trapezoidal)')
plt.axhline(y=0.52848, color='r', linestyle='--', label='Actual Integral Value (Approx)')
plt.xscale('log')  # Logarithmic scale for N values
plt.xlabel('Number of Slices (N)')
plt.ylabel('Estimated Integral I')
plt.title('Estimation of Integral I Using Trapezoidal Rule')
plt.legend()
plt.grid(True)
plt.show()
```
